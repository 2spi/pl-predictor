```python
for i in range(2101):
    name = matches_df[matches_df.columns[-1]][i]
    name = " ".join(name.split("-"))
    matches_df.replace(to_replace = matches_df[matches_df.columns[-1]][i], value = name, inplace = True)
```

```python
matches_df["date"] = pd.to_datetime(matches_df["date"])
```

```python
matches_df.dtypes
```

```python
matches_df["venue_code"] = matches_df["venue"].astype("category").cat.codes
```

```python
matches_df["opp_code"] = matches_df["opponent"].astype("category").cat.codes
```

```python
matches_df["hour"] = matches_df["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches_df["day_code"] = matches_df["date"].dt.dayofweek
```

```python
matches_df["target"] = (matches_df["result"] == "W").astype("int")
```

```python
matches_df.drop(matches_df.columns[[0]], axis = 1, inplace = True)
```

## Random Forest Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches_df[matches_df["date"] < '2023-01-01']
```

```python
test = matches_df[matches_df["date"] > '2023-01-01']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code", "xg", "xga", "poss"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
predictions = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test["target"], predictions)
```

```python
combined_df = pd.DataFrame(dict(actual=test["target"], prediction=predictions))
```

```python
pd.crosstab(index=combined_df["actual"], columns=combined_df["prediction"])
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test["target"], predictions)
```

```python
grouped_matches = matches_df.groupby("club")
```

```python
group = grouped_matches.get_group("Manchester United").sort_values("date")
```

```python
group
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches_df.groupby("club").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel("club")
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2023-01-01']
    test = data[data["date"] > '2023-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors+new_cols)
```

```python
precision
```

```python
combined = combined.merge(matches_rolling[["date", "club", "opponent", "result"]], left_index=True, right_index=True)
```

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)

combined["new_club"] = combined["club"].map(mapping)
```
